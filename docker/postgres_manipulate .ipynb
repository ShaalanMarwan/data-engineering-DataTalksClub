```python
import pandas as pd
from sqlalchemy import create_engine
pd.set_option('display.max_columns', None)
```

```python
engine = create_engine('postgresql://root:root@localhost:5433/taxi')
```

```python
engine.connect()
```

```python
query = """
SELECT 1 as number;
"""

pd.read_sql(query, con=engine)
```

```python
query = """
SELECT *
FROM public.green_taxi_data;"""
data = pd.read_sql(query, con=engine)
```

```python
query = """
SELECT * from public.green_taxi_data where lpep_pickup_datetime >= '2019-01-15'
AND lpep_pickup_datetime < '2019-01-16';
"""
green_taxi=pd.read_sql(query,con=engine)
```

```python
query = """
SELECT * FROM public.green_taxi_data
WHERE trip_distance = (
   SELECT MAX (trip_distance)
   FROM public.green_taxi_data 
);
"""
pd.read_sql(query,con=engine)
```

```python
query = """
SELECT * FROM public.green_taxi_data
WHERE passenger_count=2 and lpep_pickup_datetime >= '2019-01-01'
AND lpep_pickup_datetime < '2019-01-02';
"""
pd.read_sql(query,con=engine)
```

```python
query = """
SELECT COUNT(*) 
 FROM public.green_taxi_data
WHERE passenger_count=3 and lpep_pickup_datetime >= '2019-01-01'
AND lpep_pickup_datetime < '2019-01-02';
"""
pd.read_sql(query,con=engine)
```

```python

query = """
SELECT * FROM public.green_taxi_data
WHERE "PULocationID"=7 ORDER BY tip_amount;
"""
pd.read_sql(query,con=engine)
```
